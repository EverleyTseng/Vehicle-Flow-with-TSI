```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
video_name = '../Mask_RCNN/inferences/videos/advan_001.avi'

video_capture = cv2.VideoCapture(video_name)
count = 0
while(True):
    success, frame = video_capture.read()
    if not success:
        print('video not found')
        break
    print('frame:', count)
    cv2.imwrite('frame.png', frame)
    plt.imshow(frame)
    plt.show()
    count += 1
    break
#image = frame[350:600, 1100:1400, :]
image = frame[700:, 1400:1900, :]
cv2.imwrite('image.png', image)
plt.imshow(image)
plt.show()
```
